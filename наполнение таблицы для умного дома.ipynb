В скрипте формируется таблица с тестовыми данными для проекта "Умный дом". В таблице 10 000 строк.

Данные в таблице: дата прихода показаний, номер счетчика, тип счетчика, состояние счетчика, расширенные параметры.


```python
import pandas as pd
import numpy as np
from datetime import datetime, time, timedelta
import random
```

```python
ids = [1000,1001,1002,2000,2001,2002,3000,3001,3002,4000,4001,4002,5000,5001,5002,6000,6001,6002]
sensor_id = []
while len(sensor_id) <= 10000:
    for i in range(len(ids)):
        sensor_id.append(ids[i])
sensor_id = sensor_id[:10000]
```

```python
pd.to_datetime('2023-08-01 23:00:00')
```

```python
dates = []
ts = pd.to_datetime('2023-08-01 23:00:00')
while len(dates) <= 10000:
    for i in range(len(ids)):
        dates.append(ts)
    ts += timedelta(seconds=10)
dates = dates[:10000]
```

```python
dates = pd.to_datetime(dates)
```

```python
len(dates)
```

```python
#len(sensor_type_id)
```

```python
sensor_state = []
while len(sensor_state) <= 10000:
    sensor_state.append(random.randint(0, 1))
sensor_state = sensor_state[:10000]
```

```python
len(sensor_state)
```

```python
sensor_data = pd.DataFrame({'date':dates, 'sensor_id':sensor_id, 'sensor_state':sensor_state})
sensor_data.head()
```

```python
sensor_data.insert(2, 'sensor_type_id', 0)
sensor_data['sensor_type_id'] = (sensor_data['sensor_id']).astype('str').str[0]
sensor_data['sensor_type_id'] = sensor_data['sensor_type_id'].astype('int')
```

```python
sensor_data.info()
```

```python
sensor_data['sensor_data'] = 0
for i in range(len(sensor_data['sensor_type_id'])):
    if sensor_data['sensor_type_id'][i] == 6:
        if sensor_data['sensor_state'][i] == 1:
            sensor_data['sensor_data'][i] = str('{' + str(random.randrange(1, 4)) + ',' + str(random.randrange(1, 101)) + '}')
        else:
            sensor_data['sensor_data'][i] = {0,0}
    else:
        sensor_data['sensor_data'][i] = {}
```
